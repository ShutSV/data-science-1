## #4. Используя цикл, создать список, содержащий все целые числа от 1 до 10, и вывести его на экран.

```python
print([i for i in range(1, 11)])
```

### Ответ: сложность алгоритма - О(n), т.к. объем вычислений (количество итераций цикла) растет линейно с количеством элементов выводимого списка

## #17. Написать программу, которая принимает от пользователя строку и выводит True, если строка является палиндромом, и False в противном случае.

```python
def palindrom(item: str) -> bool:
    return True if item == item[-1::-1] else False


print(palindrom(input("Введите строку")))
```

### Ответ: сложность алгоритма - О(1), т.к. объем вычислений не зависит от длины строки, и выполняется однажды для одной строки ввода
